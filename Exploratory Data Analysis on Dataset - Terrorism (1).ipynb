# cleaning the data

```python
dataset.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region',
                             'provstate':'State','city':'City','attacktype1_txt':'Attack_type','nkill':'Killed',
                             'targsubtype1_txt':'Target_subtype','nwound':'Wounded','gname':'Group','weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
dataset=dataset[['Year','Month','Day','Country','Region','State','City','latitude','longitude','Killed','Group','Wounded','Attack_type','Motive','Target_subtype','Weapon_type']]
```

```python
dataset.head()
```

```python
#findind row of columns
dataset.shape
```

```python

#findind information
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset["Killed"]=dataset["Killed"].fillna(0)
dataset["Wounded"]=dataset["Wounded"].fillna(0)
dataset["Casualtys"]=dataset["Killed"]+dataset["Wounded"]
```

```python
dataset.describe()
```

```python
# correaltion analysis
plt.figure(figsize=(14,8))
sns.heatmap(np.round(dataset.corr(),2),annot=True,cmap="BuPu")
```

```python
#visualizing data
pd.crosstab(dataset.Year,dataset.Region).plot(kind="area",stacked=False,figsize=(14,8))
plt.title('Terrorist Activities by region in each year',fontsize=16)
plt.xlabel('Year')
plt.ylabel('No of attack')
plt.show()
```

```python
year=dataset['Year'].unique()
year_count=dataset['Year'].value_counts(dropna =False).sort_index()
plt.figure(figsize=(20,10))
sns.barplot(x=year,y=year_count)
plt.title('Attacks in years',fontsize = 20)
plt.xlabel('Attacking years',fontsize = 8)
plt.ylabel('No of attack in each year',fontsize=12)
plt.show()
```

```python

plt.subplots(figsize=(25,10))
sns.barplot(dataset['Attack_type'].value_counts()[:15].index,dataset['Attack_type'].value_counts()[:15].values,palette="rocket")
plt.title('Method of Attack',fontsize=20)
plt.xlabel('Method')
plt.ylabel('Count')
plt.show()
```

```python
plt.subplots(figsize=(15,6))
sns.barplot(dataset['Country'].value_counts()[:15].index,dataset['Country'].value_counts()[:15].values,palette='bright')
plt.title('countries with terrorist attacks',fontsize = 20)
plt.xlabel('Countries',fontsize = 10)
plt.ylabel('Count',fontsize=15)
plt.show()
```

```python
sns.barplot(dataset['Weapon_type'].value_counts()[:5].index,dataset['Weapon_type'].value_counts()[:5].values,palette='inferno')
plt.title('Top Weapon Affected by terrorism',fontsize = 10)
plt.xlabel('Weapon',fontsize =12)
plt.ylabel('Count',fontsize=12)
plt.show()
```

```python
xcasualty=dataset[["Year","Casualtys"]].groupby("Year").sum()
xcasualty.head()
```

```python
xcasualty.plot(kind="bar",color="orange",figsize=(15,6))
plt.title("Year wise casulities",fontsize=13)
plt.xlabel("Years",fontsize=13)
plt.ylabel("No of Casulities",fontsize=13)
plt.xticks(fontsize=12)
plt.show()
```

```python
xkilled=dataset[["Year","Killed"]].groupby("Year").sum()
xkilled.head()
```

```python
xwounded=dataset[["Year","Wounded"]].groupby("Year").sum()
xwounded.head()
```

```python
fig=plt.figure()
ax0=fig.add_subplot(2,1,1)
ax1=fig.add_subplot(2,1,2)
#people killed
xkilled.plot(kind="bar",color="red",figsize=(15,15),ax=ax0)
ax0.set_title("People Killed in each year")
ax0.set_xlabel("Years")
ax0.set_ylabel("No of people Killed")
#people wounded
xwounded.plot(kind="bar",color="black",figsize=(15,15),ax=ax1)
ax1.set_title("people Wounded in each year")
ax1.set_xlabel("Years")
ax1.set_ylabel("No of people Wounded")

plt.show()
```
